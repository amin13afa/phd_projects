```python
from sympy import *
import numpy as np
```

```python
α, β = symbols('α β')
w2 = Matrix([[α**2, α*β, α*β, -β**2],
             [α*β, -α**2, β**2, α*β],
             [α*β, β**2, -α**2, α*β],
             [β**2, -α*β, -α*β, -α**2]])


w = Matrix([[α, β],
            [β, -α]])
```

```python
W4 = tensorproduct(w,w,w2)
```

```python
W4
```

```python
W4_val = W4.subs({α:1/np.sqrt(2),β:1/np.sqrt(2)})
```

```python
W4_val
```
