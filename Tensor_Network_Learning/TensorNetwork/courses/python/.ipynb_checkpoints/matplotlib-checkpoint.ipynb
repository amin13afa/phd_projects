# matplotlib

```python
# import libraries 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

# Line plot

```python
plt.plot(years, iran_pop)
plt.show()
```

### scatter plot 

```python
plt.scatter(years, iran_pop)
plt.show()
```

### Histogram plot 

```python
city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1547164, 1448075, 1424641, 1249942]
```

```python
plt.hist(city_pop)
plt.show()
```

```python
plt.hist(city_pop, bins=2)   # bins app floor(sqrt(data_num))
plt.show()
```

```python
plt.hist(city_pop, bins="auto")
plt.show()
```

### Pie chart 

```python
plt.pie(city_pop, labels=city_name)
plt.show()
```

### Customization plots 

```python
plt.figure(figsize=(5,5))
plt.plot(years, iran_pop)
plt.xlabel("years")
plt.ylabel("Iran's population")
plt.xticks(years)
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['21M','28M','38M','56M',
'66M','74M','80M'])
plt.show()
```

```python
plt.scatter(np.arange(6), city_pop)
plt.xticks(range(6), ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
           ,rotation=60,fontsize='large')
plt.yticks([1000000, 2000000, 7000000],['1M','2M','7M'])
plt.show()                        
```

```python
pop_size = np.array([7153309, 2307177, 1547164, 1448075, 1424641, 1249942])/1e4
colors = ["green", "orange","blue","black", "tomato","violet"]


plt.scatter(np.arange(6), city_pop, s=pop_size, c=colors)
plt.xticks(range(6), ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
           ,rotation=60,fontsize='large')
plt.yticks([1000000, 2000000, 7000000],['1M','2M','7M'])
plt.margins(0.1)
plt.text(0,7153309, "Iran Capital", fontsize=10)
plt.show()     
```

### Multiple plot 

```python
turkey_pop = [27.47, 34.88, 43.98, 53.92, 63.24, 72.33, 79.51]

plt.plot(years, iran_pop, ls='-',marker="+", mew=8)
plt.plot(years, turkey_pop, ls='--', lw=1)
plt.title("Iran vs Turkey Population")
plt.xlabel("years")
plt.ylabel("Population")
plt.legend(["Iran", "Turkey"], loc="best")
plt.yticks([20,30,40,50,60,70,80],['20M', '30M', '40M', '50M' ,'60M', '70M', '80M',])
plt.grid()
plt.annotate('Iran-Iraq war', xytext = (1990,34.88),fontsize=12 ,xy=(1980,38.67),
             arrowprops=dict(facecolor="silver", width=4))
plt.show()
```
